```python
df = pd.read_pickle("../input/consumption_day.pkl")
meter_list = df.meter_id.unique()

print(df.shape)
print(len(meter_list))
```

```python
# last 31 days small meter reading sum is suspicious
last_cols = df.columns[-31:-1]
suspicious_a = df[df[last_cols].mean(axis=1) < 0.2][last_cols]
suspicious_a
```

```python
# last 31 days quantile 0.95, exclude outliers
df["quan"] = df[df.columns[1:]].quantile(0.95, axis=1)
for col in df.columns[1:-1]:
    df[col] = df[col] / df["quan"].values
```
